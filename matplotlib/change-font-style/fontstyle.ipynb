```python
import os
#import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
#mpl.rcdefaults()
#mpl.rcParams
plt.rcdefaults()
plt.rcParams
```

```python
#print(mpl.rcParams['font.family'], mpl.rcParams['font.sans-serif'])
print(plt.rcParams['font.family'], plt.rcParams['font.sans-serif'])
```

```python
def plot():
    x = np.arange(0, 10)
    plt.plot(x, x)
    plt.title('0123456789', fontsize=24)
    plt.xlabel('ABCDEFGHIJKLMNOPQRSTUVWXYZ', fontsize=18)
    plt.ylabel('ABCDEFGHIJKLMNOPQRSTUVWXYZ'.lower(), fontsize=18)
    plt.show()
```

```python
print(plt.rcParams['font.family'], plt.rcParams['font.sans-serif'])
fontstyle = plt.rcParams['font.sans-serif']
plot()
```

```python
#plt.rcParams.update({'font.sans-serif':'Arial'})
#plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['font.sans-serif'] = ['Arial']
print(plt.rcParams['font.family'], plt.rcParams['font.sans-serif'])
plot()
```

```python
plt.rcParams['font.family'] = 'DeJavu Serif'
plt.rcParams['font.serif'] = ['Times New Roman']
print(plt.rcParams['font.family'], plt.rcParams['font.serif'])
plot()
```
